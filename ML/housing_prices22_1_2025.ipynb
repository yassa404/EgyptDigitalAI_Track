### encoding OneHotEncoder

```python
df.head(3)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
ohe.fit(df[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']])
```

```python
encoded_data=ohe.transform(df[['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']])
```

```python
encoded_data.shape
```

```python
columns = ohe.get_feature_names_out(['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus'])
# Use encoded_data instead of encoded_array
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=columns) # Convert encoded_data to a NumPy array using toarray() method.
final_df = pd.concat([df, encoded_df], axis=1)
```

### get_dummies

```python
xx=pd.get_dummies(df,columns=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus'],drop_first=True)
```

```python
xx.head()
```

```python
xx.shape
```

### split data

```python
# split data
x=xx.drop(['price','log_price','area','log_area'],axis=1)
y=xx['log_price']
```

```python
## split to train and test
from sklearn.model_selection import train_test_split
x_tran, x_test, y_tran, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
### linear model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_tran,y_tran)
```

```python
from sklearn.metrics import mean_squared_error as mse
y_pred=model.predict(x_test)
print(mse(y_test,y_pred))
```
