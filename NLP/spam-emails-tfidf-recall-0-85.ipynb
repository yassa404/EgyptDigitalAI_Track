```python
import spacy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score , confusion_matrix, classification_report
```

```python
df=pd.read_csv('/kaggle/input/spam-emails/spam.csv')
df.head()
```

```python
df.info()
```

```python
df['Category'].value_counts()
```

```python
text=df.iloc[0,1]
text
```

```python
nlp=spacy.load('en_core_web_md')
```

```python
doc=nlp(text)
doc
```

```python
for token in doc:
    print (token)
```

```python

for token in doc:
    print(f"{token} ==> {token.lemma}")
```

```python
for token in doc:
    print(f"{token} ==> {token.lemma_}")
```

```python
for token in doc:
    if token.is_stop:
        print(f"{token} ==> Stop word")
    elif token.is_punct:
        print(f"{token} ==> is punct")
    else:
        print( f"{token} ==> Not stop")
```

```python
def prepocess(text):
    doc=nlp(text)
    filtered_text = []
    for token in doc:
        if  token.is_stop or token.is_punct:
            continue
        filtered_text.append(token.lemma_)
    return " ".join(filtered_text)
```

```python
df['PrepocessedMassage']= df['Message'].apply(prepocess)
```

```python
df
```

```python
encoder=LabelEncoder()
df['Category']=encoder.fit_transform(df['Category'])
```

```python
df.head()
```

```python
x= df['PrepocessedMassage']
y= df['Category']
```

```python
v=TfidfVectorizer()
x_v=v.fit_transform(x)
```

```python
x_v
```

```python
x
```

```python
v.vocabulary_
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_v,y,test_size=0.2,random_state=101,stratify=y)
```

## Model

```python
ir=LogisticRegression()
ir.fit(x_train,y_train)
```

```python
y_testpred=ir.score(x_train,y_train)
print(y_testpred)
```

```python
y_pred=ir.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm,annot=True,cmap='Blues')
plt.show
```

```python
print(classification_report(y_test,y_pred))
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
cmrf = confusion_matrix(y_test,y_pred)
cmrf
```

```python
sns.heatmap(cmrf,annot=True,cmap='Blues',fmt='0.3f')
```
